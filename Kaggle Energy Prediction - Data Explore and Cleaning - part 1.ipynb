### finding the timestamp time zone

```python
df_weather_train = pd.read_csv('E:/data/Kaggle Energy Prediction/weather_train.csv')
df_weather_train['timestamp'] = pd.to_datetime(df_weather_train['timestamp'])
df_weather_train['hour'] = df_weather_train['timestamp'].dt.hour
df_weather_train.head()
```

```python
# df_weather_train_study = df_weather_train[['hour', 'site_id', 'air_temperature']].groupby(['site_id','hour']).agg('mean').reset_index()
# df_weather_train_study.to_csv('E:/data/Kaggle Energy Prediction/site_temp_trend.csv')
```

#### getting hour adjustment to convert timestamp to local time

```python
df_weather_train_study['max_temp_of_the_area'] = df_weather_train_study.groupby('site_id').air_temperature.transform(np.max)
df_site_time_adj = df_weather_train_study.loc[df_weather_train_study.air_temperature == df_weather_train_study.max_temp_of_the_area].reset_index()
df_site_time_adj['time_adj'] = df_site_time_adj['hour'] - 12
df_site_time_adj
```

```python
df_train = df_train.merge(df_site_time_adj[['site_id','time_adj']], how = 'left', on = 'site_id')
df_train['hour_adj'] = np.where(df_train.hour - df_train.time_adj < 0 ,24 + df_train.hour - df_train.time_adj ,df_train.hour - df_train.time_adj)
df_train.head()
```

```python
pd.pivot_table(df_train , values='meter_reading', index=['meter', 'hour_adj'],columns=['site_id'], aggfunc=np.mean)
```
